```python
# Environment: dash1_8_0_env
import pandas as pd     #(version 1.0.0)
import plotly           #(version 4.5.0)
import plotly.express as px
import plotly.io as pio
import glob
import os

files = os.path.join("C:\\Users\\xyz\\Documents\\Python Scripts\\treasury_rate\\", "daily-treasury-rates_*.csv")

files = glob.glob(files)

#print (os.getcwd()) # Prints the working directory

df = pd.concat(map(pd.read_csv, files), ignore_index=True) 
df.drop('2 Mo', axis=1, inplace=True)  #2 Month yield is not avaliable historically, so dropping the column
df.head()
```

```python
#df = pd.read_csv("Daily_Treasury_MainData.csv")


df['Date']= pd.to_datetime(df['Date'])

#df.head()
#df = df.sort_values(by='Date',ascending=True)
#df['Date'].head()
#df_sub = df.loc[df['Date'] == '1/10/2022']
#df_sub.head()
```

```python
df_melted = df.melt('Date',var_name='variable',value_name='value')
#df_melted.head()


```

```python
from pandas.api.types import CategoricalDtype

cat_size_order = CategoricalDtype(["1 Mo","3 Mo","6 Mo","1 Yr","2 Yr","3 Yr","5 Yr","7 Yr","10 Yr","20 Yr","30 Yr"], 
    ordered=True
)

df_melted['variable'] = df_melted['variable'].astype(cat_size_order)

#df_melted['variable'] 
```

```python
df_melted = df_melted.sort_values(by= ['Date','variable'], ascending=True)

df_melted['Date'] = df_melted['Date'].astype(str)
```

```python
df_melted.head(100)
```

```python
fig = px.line(df_melted, x="variable", y="value", title='Treasury Rate curve', markers=True, animation_frame="Date"
          ,labels={"variable": "Term","value": "% Rate"          
                 }, range_y=[-0.1,5.0])

#fig.update_yaxes(tickvals=["1 Mo","2 Mo","3 Mo","6 Mo","1 Yr","2 Yr","3 Yr","5 Yr","7 Yr","10 Yr","20 Yr","30 Yr"])

fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 100
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 10

#pio.show(fig)

fig.show()
```
